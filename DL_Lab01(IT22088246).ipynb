1. Create a random 4x4 array from an exponential distribution

```python
import numpy as np
array_4x4 = np.random.exponential(scale=1.0, size=(4, 4))
print(array_4x4)
```

 2. Create a 100000x1 array and plot histograms

```python
import matplotlib.pyplot as plt

exp_data = np.random.exponential(scale=1.0, size=100000)
uniform_data = np.random.rand(100000)
normal_data = np.random.randn(100000)

plt.hist(exp_data, density=True, bins=100, histtype='step', color='red', label='Exponential')
plt.hist(uniform_data, density=True, bins=100, histtype='step', color='blue', label='Uniform')
plt.hist(normal_data, density=True, bins=100, histtype='step', color='green', label='Normal')

plt.title("Random Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend(loc='upper right')
plt.axis([-2.5, 8, 0, 1.1])
plt.show()
```

3. 3D Plot of Z = X² + Y²

```python
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_title("Z = X² + Y²")
plt.show()
```

4. Pearson and Spearman Correlation with Heatmap

```python
import seaborn as sns
import pandas as pd

# Load your Pokémon dataset
df = pd.read_csv('pokemon.csv')

# Select the features
features = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
selected = df[features]

# Pearson Correlation
pearson_corr = selected.corr(method='pearson')
spearman_corr = selected.corr(method='spearman')

# Heatmap for Pearson
plt.figure(figsize=(8, 6))
sns.heatmap(pearson_corr, annot=True, cmap='coolwarm')
plt.title('Pearson Correlation Heatmap')
plt.show()

# Heatmap for Spearman
plt.figure(figsize=(8, 6))
sns.heatmap(spearman_corr, annot=True, cmap='magma')
plt.title('Spearman Correlation Heatmap')
plt.show()
```

# Task 02

1. Load Data and Filter Specific Columns

```python
df = pd.read_csv('/content/Au_nanoparticle_dataset.csv')
new_df = df[['N_total', 'N_bulk', 'N_surface', 'R_avg']]
```

2. Display First 20 Samples

```python
print(new_df.head(20))
```

3. Mean, Std, Quartiles

```python
print("Mean:\n", new_df.mean())
print("\nStandard Deviation:\n", new_df.std())
print("\nQuartiles:\n", new_df.quantile([0.25, 0.5, 0.75]))
```

4. Histogram in 1x4 Layout

```python
fig, axes = plt.subplots(1, 4, figsize=(20, 4))
features = ['N_total', 'N_bulk', 'N_surface', 'R_avg']

for i, col in enumerate(features):
    axes[i].hist(new_df[col], bins=30, color='skyblue', edgecolor='black')
    axes[i].set_title(f'Histogram of {col}')
plt.tight_layout()
plt.show()
```

5. Seaborn Pairplot (basic)

```python
sns.pairplot(new_df)
plt.show()
```

6. Pairplot Customization with KDE

```python
g = sns.PairGrid(new_df)
g.map_upper(sns.histplot)
g.map_diag(sns.histplot, kde=True)
g.map_lower(sns.kdeplot)
plt.show()
```
